# Bivariate Analysis - Complete Guide

## What is Bivariate Analysis?
Bivariate analysis involves analyzing **two variables simultaneously** to understand the relationship between them. This helps us discover:
- **Patterns**: How one variable changes with respect to another
- **Correlations**: Whether variables are positively or negatively related
- **Trends**: Directional movements over time or categories
- **Distributions**: How data is spread across different groups

## Types of Bivariate Relationships:
1. **Numerical - Numerical**: Both variables are continuous (e.g., height vs weight)
2. **Categorical - Numerical**: One category, one number (e.g., gender vs salary)
3. **Categorical - Categorical**: Both are categories (e.g., gender vs smoking status)

## 1. Scatterplot (Numerical - Numerical)

### What is a Scatterplot?
A scatterplot shows the relationship between two numerical variables by plotting points on a 2D plane.

### When to use?
- When both variables are **continuous/numerical**
- To detect **linear or non-linear relationships**
- To identify **outliers** and **clusters**
- To check **correlation** between variables

### How it helps understand data:
- **Positive Correlation**: Points trend upward (one variable increases with the other)
- **Negative Correlation**: Points trend downward (one variable decreases as other increases)
- **No Correlation**: Points scattered randomly
- **Strength of relationship**: How closely points follow a pattern

### Basic Scatterplot
**Function**: `sns.scatterplot(x, y, data)`
- **x**: Column name for x-axis (independent variable)
- **y**: Column name for y-axis (dependent variable)
- **data**: DataFrame containing the data


```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
```

### Scatterplot with Color Coding (hue parameter)

**New Parameter**: `hue='sex'`
- **Purpose**: Colors points based on a third categorical variable
- **Benefit**: Reveals patterns within subgroups

**Data Understanding**:
- See if males and females tip differently for same bill amount
- Identify if there are distinct tipping patterns by gender
- Check if one group has higher variance than another


```python
sns.scatterplot(x='total_bill', y='tip', hue='sex', data=tips)
```

### Scatterplot with Multiple Encodings (hue + style)

**New Parameter**: `style='smoker'`
- **Purpose**: Uses different marker shapes for categories
- **Benefit**: Adds another dimension to the plot without cluttering

**Combined Understanding**:
- **hue (color)** = sex → Shows gender differences
- **style (marker shape)** = smoker → Shows smoking status differences
- Can now analyze 4 groups simultaneously: (Male-Smoker, Male-NonSmoker, Female-Smoker, Female-NonSmoker)


```python
sns.scatterplot(x='total_bill', y='tip', hue='sex', style='smoker', data=tips)
```

### Scatterplot with Size Encoding (hue + style + size)

**New Parameter**: `size='size'`
- **Purpose**: Varies point size based on a numerical variable (party size in this case)
- **Benefit**: Adds 4th dimension to visualization

**Complete Understanding**:
- **x-axis**: Total bill amount
- **y-axis**: Tip amount
- **Color (hue)**: Gender
- **Shape (style)**: Smoking status
- **Size**: Party size (number of people)

**Insights we can gain**:
- Do larger parties tip more/less proportionally?
- How do gender, smoking status, and party size interact with tipping behavior?
- Are there distinct patterns in how different groups tip?


```python
sns.scatterplot(x='total_bill', y='tip', hue='sex', style='smoker', size='size', data=tips)
```

## 3. Box Plot (Numerical - Categorical)

### What is a Box Plot?
A box plot (box-and-whisker plot) shows the **distribution** of a numerical variable across different categories, displaying the **five-number summary**:
1. **Minimum** (lower whisker)
2. **Q1** (25th percentile - bottom of box)
3. **Median** (50th percentile - line in box)
4. **Q3** (75th percentile - top of box)
5. **Maximum** (upper whisker)
6. **Outliers** (individual points beyond whiskers)

### When to use?
- To compare **distributions** across categories (not just averages)
- To identify **outliers** in different groups
- To understand **spread/variance** within groups
- To see **skewness** (is median closer to Q1 or Q3?)

### Box Plot vs Bar Plot:
- **Bar plot**: Shows only the **mean** (single value)
- **Box plot**: Shows the entire **distribution** (spread, quartiles, outliers)

### How it helps understand data:
- **Compare medians**: Central tendency across groups
- **Compare spreads**: Which group has more variability?
- **Identify outliers**: Unusual values that might need investigation
- **Skewness**: Is data symmetric or skewed?
- **IQR (Interquartile Range)**: Q3 - Q1 = height of box = middle 50% of data


### Box Plot Example: Age Distribution by Gender and Survival

**Function**: `sns.boxplot(x, y, hue, data)`
- **x='Sex'**: Categorical variable (Male/Female)
- **y='Age'**: Numerical variable to show distribution
- **hue='Survived'**: Additional categorical split (0=died, 1=survived)

**Understanding from this plot**:
- **Age distribution patterns**: Were survivors younger/older than non-survivors?
- **Gender differences**: How did age distribution differ between males and females?
- **Outliers**: Any unusually young/old passengers in each group?
- **Spread comparison**: Did one group have more age variability?
- **Median differences**: Central age in each group

**Key Insights to look for**:
- Women and children first policy → Female survivors might be younger
- Male survivors might be older/wealthier (could afford better cabins)
- Outliers might represent notable individuals (very young children, elderly passengers)


```python
sns.boxplot(x='Sex', y='Age', hue='Survived', data=titanic)
```

## 4. Distribution Plot / Histogram with KDE (Numerical - Categorical)

### What is a Distribution Plot?
Shows the **probability distribution** of a numerical variable, combining:
- **Histogram**: Bars showing frequency of values in bins
- **KDE (Kernel Density Estimate)**: Smooth curve representing the probability density

### When to use?
- To understand the **shape** of data distribution
- To check if data is **normally distributed**
- To identify **multiple peaks** (bimodal/multimodal distributions)
- To compare distributions of different groups

### How it helps understand data:
- **Central tendency**: Where most values cluster
- **Spread**: How dispersed the data is
- **Skewness**: Is the tail longer on left (negative skew) or right (positive skew)?
- **Modality**: One peak (unimodal), two peaks (bimodal), many peaks (multimodal)
- **Outliers**: Values far from the main distribution

### KDE (Kernel Density Estimation):
- Creates a **smooth curve** instead of discrete bars
- Shows **probability density** (area under curve = 1)
- Makes it easier to **compare multiple distributions**
- **kde=True** parameter adds this smooth curve on top of histogram


### Distribution Plot Example: Age Distribution

**Function**: `sns.histplot(data, kde=True)`
- **First parameter**: Column data to plot
- **kde=True**: Adds the smooth Kernel Density Estimate curve

**Note**: Previously `sns.distplot()` was used but it's now deprecated. Use `histplot()` instead.

**Understanding from this plot**:
- **Shape**: Is age normally distributed or skewed?
- **Common ages**: Which age groups were most common on Titanic?
- **Gaps**: Any age groups notably absent?
- **Peaks**: Most frequent age ranges


```python
sns.histplot(titanic['Age'], kde=True)
# sns.distplot(titanic['Age'])  # Old deprecated method
```

### Comparing Distributions with KDE: Survived vs Not Survived

**Function**: `sns.kdeplot(data)`
- **Purpose**: Plot only the smooth KDE curve (no histogram bars)
- **Benefit**: Clean comparison of multiple distributions on same plot

**Code Breakdown**:
- `titanic[titanic['Survived']==0]['Age']`: Filter data where Survived=0 (died)
- `titanic[titanic['Survived']==1]['Age']`: Filter data where Survived=1 (survived)
- Two separate `kdeplot` calls create overlapping curves

**Understanding from this plot**:
- **Peak differences**: Were survivors concentrated in different age ranges?
- **Shape differences**: Different distribution shapes for survivors vs non-survivors?
- **Overlap**: Where distributions overlap vs where they differ
- **Key insight**: If survivor curve peaks lower (younger ages), validates "women and children first"

**Better approach**: Could use `hue='Survived'` in single `kdeplot()` call for automatic coloring and legend


```python
sns.kdeplot(titanic[titanic['Survived']==0]['Age'])
sns.kdeplot(titanic[titanic['Survived']==1]['Age'])
```

## 5. HeatMap (Categorical - Categorical)

### What is a Heatmap?
A heatmap displays **numerical values as colors** in a matrix format. In categorical analysis, it shows the **frequency/count** of combinations between two categorical variables.

### When to use?
- To analyze relationship between **two categorical variables**
- To show **counts or percentages** in a crosstab format
- To identify **patterns** in categorical data
- To visualize **contingency tables**

### How it helps understand data:
- **Color intensity**: Darker/lighter colors show higher/lower frequencies
- **Patterns**: Diagonal patterns, clusters, or sparse regions
- **Quick comparison**: Visual scanning instead of reading numbers
- **Association**: Strong colors in certain cells indicate association between categories

### Key Function: pd.crosstab()
**Purpose**: Creates a contingency table (frequency table) showing counts of combinations
- **Syntax**: `pd.crosstab(index_variable, column_variable)`
- **Output**: A table with counts of each combination
- **Example**: crosstab(Pclass, Survived) shows how many people in each class survived/died


### Heatmap Example: Survival by Passenger Class

**Function**: `sns.heatmap(data)`
- **Input**: A 2D matrix (from pd.crosstab)
- **Default parameters**:
  - annot=False (no numbers in cells)
  - cmap='viridis' (default color scheme)

**Code Breakdown**:
- `pd.crosstab(titanic['Pclass'], titanic['Survived'])`: Creates frequency table
  - Rows: Passenger classes (1, 2, 3)
  - Columns: Survival status (0=died, 1=survived)
  - Values: Count of passengers in each combination

**Understanding from this plot**:
- **Class-based survival**: Did 1st class passengers have better survival rates?
- **Raw counts**: How many people in each class survived vs died?
- **Pattern**: If color is lighter for class 1, column 1 (survived), shows better survival

**Limitations**: Shows counts, not percentages. High counts in 3rd class might just be because there were more 3rd class passengers.

**Enhancement**: Add `annot=True` to show numbers in cells


```python
sns.heatmap(pd.crosstab(titanic['Pclass'], titanic['Survived']))
```

### Calculating Survival Rates by Port of Embarkation

**Function Chain**: `groupby().mean()`

**Step-by-step breakdown**:
1. `titanic.groupby('Embarked')`: Group passengers by embarkation port
   - **C** = Cherbourg, **Q** = Queenstown, **S** = Southampton
2. `['Survived']`: Select only the Survived column
3. `.mean()`: Calculate average of Survived (0s and 1s)
   - Since Survived is 0 or 1, mean gives survival rate (percentage as decimal)
4. `* 100`: Convert to percentage

**Understanding from this result**:
- **Survival percentage** by port: Which port had better survival rates?
- **Possible reasons for differences**:
  - Different class distributions from each port
  - Cherbourg (C) might have more 1st class passengers
  - Southampton (S) was the main port, more 3rd class passengers

**Why this matters**:
- Shows that where you boarded might correlate with survival
- Helps identify indirect relationships (port → class → survival)
- Can inform further analysis about passenger demographics


```python
titanic.groupby('Embarked')['Survived'].mean() * 100
```

## 6. ClusterMap (Categorical - Categorical)

### What is a ClusterMap?
A clustermap is an **enhanced heatmap** that automatically:
1. **Reorders rows and columns** based on similarity (hierarchical clustering)
2. Adds **dendrograms** (tree diagrams) showing how rows/columns are grouped
3. Groups **similar patterns** together

### When to use?
- When you want to **discover hidden patterns** in categorical data
- To find **natural groupings** without pre-defining them
- To see which categories **behave similarly**
- When dealing with **many categories** (hard to see patterns in regular heatmap)

### How it helps understand data:
- **Automatic pattern discovery**: Algorithm finds similarities you might miss
- **Dendrograms**: Show hierarchy of relationships (which categories cluster together)
- **Reordering**: Places similar rows/columns next to each other
- **Better insights**: Reveals structure in complex categorical relationships

### Difference from Heatmap:
- **Heatmap**: Shows data in original order
- **Clustermap**: Reorganizes data to reveal patterns

### Hierarchical Clustering:
- Groups data points into a tree-like structure
- Similar items grouped together first, then merged into larger clusters
- Dendrogram height shows how different groups are


### ClusterMap Example: Survival by Number of Parents/Children

**Function**: `sns.clustermap(data)`
- **Input**: 2D matrix (from pd.crosstab)
- **Process**: Automatically clusters rows and columns

**Code Breakdown**:
- `pd.crosstab(titanic['Parch'], titanic['Survived'])`:
  - **Parch**: Number of parents/children aboard
  - **Survived**: 0 or 1
  - Creates frequency table

**Understanding from this plot**:
- **Family size patterns**: How did having family members affect survival?
- **Dendrograms**: Show which Parch values (family sizes) had similar survival patterns
- **Clusters**: Groups of family sizes with similar outcomes
- **Possible insights**:
  - Small families (1-3 members) might cluster together with good survival
  - Traveling alone (Parch=0) might have different pattern
  - Very large families might struggle to survive together

**Why clustering matters here**:
- Original order (0,1,2,3...) might hide patterns
- Clustering reveals which family sizes behaved similarly
- Helps identify optimal family size for survival


```python
sns.clustermap(pd.crosstab(titanic['Parch'], titanic['Survived']))
```

## 7. Pairplot (Multi-variable Numerical Analysis)

### What is a Pairplot?
A pairplot creates a **grid of plots** showing:
- **Scatterplots**: For every pair of numerical variables (off-diagonal)
- **Distribution plots**: For each variable against itself (diagonal)

### When to use?
- When you have **multiple numerical features** to analyze
- To explore **all pairwise relationships** at once
- To identify **correlated variables**
- To see how features **separate different classes** (when using hue)
- In **exploratory data analysis** before building models

### How it helps understand data:
- **Comprehensive view**: See all relationships in one figure
- **Correlation detection**: Which variables move together?
- **Feature selection**: Which features best separate classes?
- **Outlier detection**: Unusual patterns visible across multiple views
- **Class separation**: When using hue, see which feature pairs best distinguish classes

### Structure:
- **N features** → **N×N grid** of plots
- **Diagonal**: Histograms/KDE of individual variables
- **Off-diagonal**: Scatterplots of variable pairs
- **Symmetry**: Upper and lower triangles are mirror images (same data, axes flipped)

### For Machine Learning:
- Helps identify which features are most predictive
- Shows if features are redundant (highly correlated)
- Reveals if classes are linearly separable

## 8. Line Plot (Numerical - Numerical, especially Time Series)

### What is a Line Plot?
A line plot shows the relationship between two numerical variables by **connecting points with lines**, typically used for:
- **Time series data**: How values change over time
- **Trends**: Overall direction (increasing, decreasing, stable)
- **Continuous relationships**: When x-axis represents ordered/sequential data

### When to use?
- **Time series**: Date/time on x-axis
- **Sequential data**: When order matters (years, stages, steps)
- **Trend analysis**: To show direction and rate of change
- **Comparing multiple series**: Using hue to show different groups

### How it helps understand data:
- **Trends**: Upward, downward, or cyclical patterns
- **Seasonality**: Repeating patterns (e.g., monthly/yearly cycles)
- **Volatility**: How much values fluctuate
- **Anomalies**: Sudden spikes or drops
- **Growth rate**: Steepness of line indicates speed of change

### Line Plot vs Scatter Plot:
- **Scatter**: Points represent individual observations, no assumption of order
- **Line**: Connects points in sequence, emphasizes continuity and trend


```python
flights.head()
```

### Data Preparation for Line Plot: Aggregating Yearly Passengers

**Goal**: Convert monthly data to yearly totals for clearer trend visualization

**Function Chain**: `groupby().sum().reset_index()`

**Step-by-step breakdown**:
1. `flights.groupby('year')`: Group data by year
   - All months of 1949 together, all months of 1950 together, etc.

2. `.sum(numeric_only=True)`: Sum numerical columns for each year
   - **numeric_only=True**: Only sum numerical columns (passengers), ignore 'month'
   - Adds up all 12 monthly passenger counts per year

3. `.reset_index()`: Convert year from index back to regular column
   - Makes data suitable for plotting (need year as a column, not index)
   - Creates clean DataFrame with 'year' and 'passengers' columns

**Why this transformation?**
- Original data has 144 rows (12 months × 12 years) → too granular for trend
- Aggregated data has 12 rows (one per year) → clearer long-term trend
- Removes monthly seasonality, shows overall growth

**Understanding**:
- This is **data aggregation**: combining multiple values into summary statistics
- Common preprocessing step for time series visualization
- Allows focus on big picture rather than month-to-month fluctuations


```python
new = flights.groupby('year').sum(numeric_only=True).reset_index()
```

### Line Plot Example: Yearly Passenger Trend

**Function**: `sns.lineplot(x, y, data)`
- **x='year'**: Sequential time variable
- **y='passengers'**: Numerical value to track
- **data=new**: Aggregated yearly data

**Understanding from this plot**:
1. **Overall trend**: Is air travel increasing over time?
   - **Expected**: Upward trend from 1949-1960 (post-WWII growth in aviation)

2. **Growth rate**:
   - **Steep line**: Rapid growth
   - **Gentle slope**: Slow growth
   - **Curve shape**: Linear vs exponential growth

3. **Patterns**:
   - **Consistent growth**: Smooth upward line
   - **Plateaus**: Periods of no growth
   - **Dips**: Years with decreased travel (economic issues, events)

4. **Business insights**:
   - Informs capacity planning for airlines
   - Shows market expansion
   - Can project future demand

**Historical context**: 
- 1949-1960 saw rapid aviation growth
- Commercial jet age beginning
- Rising middle class with disposable income for travel


```python
sns.lineplot(x='year', y='passengers', data=new)
```

## 9. Heatmap with Pivot Table (Time Series Pattern Analysis)

### What is a Pivot Table?
A pivot table **reshapes data** from long format to wide format, creating a matrix where:
- **Rows**: One categorical variable (e.g., months)
- **Columns**: Another categorical variable (e.g., years)
- **Values**: Aggregated numerical data (e.g., sum, mean, count)

### When to use Pivot Table + Heatmap?
- **Time series with multiple dimensions**: Month + Year, Day + Hour, etc.
- **Seasonal pattern detection**: How values vary across recurring periods
- **Comparing multiple time periods**: Years, quarters, weeks
- **Spot anomalies**: Unusual months/years stand out in color

### How it helps understand data:
1. **Seasonality**: Which months are consistently high/low across years?
2. **Year-over-year comparison**: Did summer 1950 have more passengers than summer 1949?
3. **Trend patterns**: Growth across both dimensions visible
4. **Outliers**: Unusual combinations stand out
5. **Business cycles**: Identify peak/off-peak periods


### Heatmap Example: Monthly Passenger Patterns Across Years

**Function**: `flights.pivot_table(values, index, columns, observed)`

**Parameters explained**:
- **values='passengers'**: What to fill the cells with (passenger counts)
- **index='month'**: Rows of the pivot table (12 months)
- **columns='year'**: Columns of the pivot table (years 1949-1960)
- **observed=False**: Include all possible categories even if not in data

**Result**: 12×12 matrix (12 months × 12 years) where each cell = passenger count for that month-year combination

**Then wrapped in sns.heatmap()** to visualize the matrix:
- **Light colors**: High passenger counts
- **Dark colors**: Low passenger counts

**Understanding from this plot**:
1. **Seasonal patterns**:
   - **Summer months** (Jun, Jul, Aug): Likely lighter (more passengers)
   - **Winter months** (Jan, Feb): Likely darker (fewer passengers)
   - Pattern should repeat across columns (years)

2. **Yearly growth**:
   - **Horizontal trend**: Each row should get lighter left to right
   - Shows passenger increase year over year

3. **Combined insights**:
   - **Top right corner** (recent summer): Highest values
   - **Bottom left corner** (early winter): Lowest values
   - Diagonal gradient showing growth + seasonality

4. **Business value**:
   - **Staffing decisions**: Hire more in summer months
   - **Pricing strategy**: Higher fares in peak months
   - **Maintenance scheduling**: Do it in low-traffic months


```python
sns.heatmap(flights.pivot_table(values='passengers', index='month', columns='year', observed=False))
```

### ClusterMap Example: Finding Similar Months and Years

**Function**: `sns.clustermap(pivot_table)`
- **Input**: Same pivot table as above
- **Process**: Hierarchical clustering on both rows (months) and columns (years)

**Understanding from this plot**:
1. **Month clustering (rows)**:
   - **Similar months grouped together**: 
     - Summer months (Jun, Jul, Aug) will cluster
     - Winter months (Dec, Jan, Feb) will cluster
   - **Dendrogram shows**: Which months have similar passenger patterns across years

2. **Year clustering (columns)**:
   - **Similar years grouped together**:
     - Early years (1949-1951) might cluster (lower overall traffic)
     - Later years (1958-1960) might cluster (higher overall traffic)
   - Shows eras of similar travel patterns

3. **Why clustering is valuable here**:
   - **Original order** (Jan-Dec, 1949-1960) is chronological
   - **Clustered order** reveals behavioral similarities
   - Example: May and September might cluster (shoulder seasons)

4. **Business insights**:
   - **Months in same cluster**: Can use same marketing/pricing strategy
   - **Years in same cluster**: Historical trends, identify turning points
   - **Pattern discovery**: Find non-obvious similarities (e.g., "September behaves like May")

**Difference from regular heatmap**:
- **Heatmap**: Shows data in calendar order (easy to find specific month-year)
- **Clustermap**: Reorganizes to show patterns (easy to find similar periods)

# SUMMARY: When to Use Each Visualization

## Quick Reference Guide:

| Visualization | Variable Types | Best For | Key Insight |
|--------------|---------------|----------|-------------|
| **Scatterplot** | Numerical-Numerical | Correlation, trends, outliers | How two continuous variables relate |
| **Bar Plot** | Categorical-Numerical | Comparing averages across groups | Which category has highest/lowest average |
| **Box Plot** | Categorical-Numerical | Comparing distributions, outliers | Spread, quartiles, and outliers by group |
| **Histogram/KDE** | Numerical (+optional Categorical) | Distribution shape, normality | How data is distributed, skewness |
| **Heatmap** | Categorical-Categorical | Frequency patterns, associations | Count/frequency of category combinations |
| **ClusterMap** | Categorical-Categorical | Hidden patterns, groupings | Natural clusters in categorical data |
| **Pairplot** | Multiple Numerical | All pairwise relationships | Comprehensive feature analysis |
| **Line Plot** | Numerical-Numerical (time) | Trends over time | Growth, seasonality, cycles |
