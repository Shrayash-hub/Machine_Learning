This cell imports two essential libraries:

* **`requests`**: A library for making HTTP requests to fetch data from web APIs.

```python
import pandas as pd
import requests
```

makes an HTTP GET request to The Movie Database (TMDb) API to retrieve a list of top-rated movies. It uses:

*   **`requests.get()`**: To send the GET request.
*   **`api_key`**: Your unique API key for authentication with TMDb.
*   **`language=en-US`**: Specifies the language for the results.
*   **`page=1`**: Requests the first page of results.

```python
response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=0afd6c2e672c77fc358fdbaa0d3d86a5&language=en-US&page=1')
```

This cell processes the API response and creates an initial pandas DataFrame:

*   **`response.json()`**: Parses the JSON content of the API response into a Python dictionary or list.
*   **`['results']`**: Accesses the list of movie data from the JSON response.
*   **`pd.DataFrame()`**: Converts this list of movie data into a pandas DataFrame.
*   **`[['id','title',...,'vote_count']]`**: Selects specific columns to include in the DataFrame, effectively filtering the raw API data.

```python
temp_df = pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
```

```python
temp_df.head()
```

This cell initializes an empty pandas DataFrame named `df`. This is often done when you plan to populate the DataFrame iteratively, for example, by concatenating data from multiple sources in a loop.

```python
df = pd.DataFrame()
```

```python
df
```

This crucial cell iterates through multiple pages of the TMDb API to collect a larger dataset:

*   **`for i in range(1,429)`**: Loops from page 1 up to (but not including) page 429.
*   **`requests.get(...)`**: Makes an API call for each page, dynamically inserting the page number `i` into the URL.
*   **`pd.DataFrame(...)`**: Creates a temporary DataFrame (`temprory_df`) for the results of each page.
*   **`pd.concat([df, temprory_df], ignore_index=True)`**: Appends the `temprory_df` from each page to the main `df` DataFrame. `ignore_index=True` ensures the new DataFrame has a continuous index rather than retaining the original indices from `df` and `temprory_df`.

```python
for i in range(1,429):
  response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=0afd6c2e672c77fc358fdbaa0d3d86a5&language=en-US&page={}'.format(i))
  temprory_df = pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
  df = pd.concat([df, temprory_df], ignore_index=True)
```

The **`df.to_csv('movies.csv')`** function saves the entire `df` DataFrame to a CSV (Comma Separated Values) file named `movies.csv`. This allows you to persist the collected data for later use or sharing.

```python
df.to_csv('movies.csv')
```
