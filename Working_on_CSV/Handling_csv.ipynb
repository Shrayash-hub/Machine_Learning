# Opening a csv file from a url

```python
import requests # -> most popular library to fetch data from internet
from io import StringIO # -> it saves our text to the RAM in a virtual file

url = "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv"

# when we connect from a website , so our browser send some info , which includes user-agent -> it tells
# to the server that request is sended by whom , some websites block simple scripts , adding user-agent
# header makes your request look like a real browser , that reduces the chances of blocking
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}

req = requests.get(url, headers=headers) # get info request

data = StringIO(req.text) # in req.text , all the data exctracted is saved in a single string format
# by giving it to StringIO -> our data became a file like object
pd.read_csv(data)
```

# Advanced Reading: pd.read_csv() : with parameters

# 1. sep ( or delimiter):
when values are separated with ';' or '\t' rather than commas

```python
pd.read_csv('movie_titles_metadata.tsv',sep='\t')
```

# 2. header :
 when there is no clm name header in our file , or more than one header rows

```python
# when there is no header
pd.read_csv('movie_titles_metadata.tsv',sep='\t',header=None) # header = 1,2 we can also give
```

```python
# giving your own clm names
pd.read_csv('movie_titles_metadata.tsv',sep='\t',names=['sno','name','release_year','rating','votes','genres'])
```

# 3. index_col :
when we want to make any clm of a file as the index of the dataframe

```python
pd.read_csv('aug_train.csv',index_col='city') # city is going to be at first index
```

# 4 . use_cols :
if we want limited number of clm , from the large dataset file

```python
pd.read_csv('aug_train.csv',usecols=['enrollee_id','gender','education_level'])
```

# 4.1 Squeeze parameter

-> The squeeze parameter in pd.read_csv() is no longer supported in newer versions of pandas. To fix this, I'll remove the squeeze parameter and select the 'gender' column directly after loading the CSV, which will return a pandas Series.

```python
pd.read_csv('aug_train.csv',usecols=['gender'])['gender']
```

# 5. skiprows/nrows

-> skiprows: ignore some starting rows from the file
-> nrows: read only given number of rows

```python
pd.read_csv('aug_train.csv',nrows=15)
```

# 6. encoding:
it gives a very common error , when our file is not in utf-8 format , if we get the UnicodeDecodeError , we can try latin1 , cp1252

```python
pd.read_csv('zomato.csv',encoding='latin-1')
```

# 7. on_bad_lines:
This parameter controls what happens when a "bad line" (a row with too many or too few columns) is encountered. i.e for corrupted rows

it accepts three string values:

'error' (default): This will raise an exception (an error) and stop the program if a bad line is found.

'warn': This will skip the bad line and show a warning message, but the rest of the file will still be processed.

'skip': This will silently skip over the bad lines without giving you any notification.

```python
pd.read_csv('BX-Books.csv', sep=';', encoding="latin-1",on_bad_lines='warn') # -> warning msgs are shown
```

# 10. converters:
allows you to apply a custom function to one or more columns as the data is being read

Sometimes, a column needs more than just a type change. You might need to strip currency symbols, remove percentage signs, or perform a custom calculation.

You pass a dictionary where the keys are the column names (or indices) and the values are the functions you want to apply to them.

```python
def rename(name):
    if name == "Royal Challengers Bangalore":
        return "RCB"
    else:
        return name
```

```python
rename("Royal Challengers Bangalore")
```

```python
pd.read_csv('IPL Matches 2008-2020.csv',converters={'team1':rename})
```

# 11. parse_dates:
This converts them to the proper datetime format during the loading process itself.

```python
pd.read_csv('IPL Matches 2008-2020.csv',parse_dates=['date']).info()
```

syntex : df = pd.read_csv(

    keep_date_col=True  # Optional: keeps the original year, month, day columns
)

# 12. Loading hue dataset in chunks:
Jab CSV file aapke system ki RAM se badi ho (e.g., 10GB file aur 8GB RAM), tab pd.read_csv() fail ho jayega.

```python
dfs = pd.read_csv('aug_train.csv',chunksize=5000)

total_rows = 0
for chunk in dfs:
    # Yahan har 'chunk' ek chhota DataFrame hai (5000 rows ka)
    # Aap is chunk par koi bhi operation kar sakte hain
    total_rows += len(chunk)

print(f"Total rows in the file: {total_rows}")
```
