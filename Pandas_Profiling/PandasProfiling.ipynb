# 📊 Automated EDA with `ydata-profiling`

## What is Automated EDA?

Manual EDA means writing dozens of lines — `df.head()`, `df.info()`, `df.describe()`, multiple plots — just to understand one dataset.

**Automated EDA** does all of that in a single command and generates a complete, interactive HTML report.

> 🎯 **Goal of this notebook:** Use `ydata-profiling` (formerly `pandas-profiling`) to generate a full EDA report for the Titanic dataset in just ~3 lines of code.

---

### 📋 What `ydata-profiling` gives you automatically:

| Section in Report | What it shows |
|-------------------|--------------|
| **Overview** | Dataset shape, missing values %, duplicate rows, memory usage |
| **Variables** | Per-column stats: type, unique values, missing count, distribution |
| **Interactions** | Scatter plots between pairs of numerical columns |
| **Correlations** | Pearson, Spearman, Kendall heatmaps |
| **Missing Values** | Heatmap + bar chart of where NaNs are |
| **Duplicates** | Shows duplicate rows if any |
| **Sample** | First and last few rows |

All of this generated from **one function call** — instead of 50+ lines of manual code.

---



### Why we need it before profiling:
`ydata-profiling` works by taking an existing **pandas DataFrame as input** — it doesn't load data itself. So the workflow is always:
```
Load data with pandas  →  Pass DataFrame to ProfileReport  →  Get HTML report
```

## 🚀 Generating the Full EDA Report

```python
from ydata_profiling import ProfileReport
profile = ProfileReport(df, explorative=True)
profile.to_file("output.html")
```

This is the **core cell** of the entire notebook — 3 lines replace hundreds of lines of manual EDA code.

---

### Line 1 — Importing `ProfileReport`

```python
from ydata_profiling import ProfileReport
```

- `ydata_profiling` is the **package** (installed above).
- `ProfileReport` is the **main class** inside it — the only thing you need to import.
- It takes your DataFrame and handles all analysis internally.

---

### Line 2 — Creating the Profile

```python
profile = ProfileReport(df, explorative=True)
```

#### What `ProfileReport(df, ...)` does:
- Takes your DataFrame `df` as input.
- Scans every column, computes statistics, detects types, finds correlations, identifies missing values, detects duplicates.
- Stores the complete analysis in the `profile` object.

#### Key parameter — `explorative=True`:

| Setting | Behaviour |
|---------|-----------|
| `explorative=False` (default) | Faster, conservative — skips some expensive computations. Good for large datasets. |
| `explorative=True` | **Full analysis** — enables additional plots, more correlation methods, interaction plots. Recommended for EDA. |

#### Other useful parameters:

```python
# Minimal report (very fast, for large datasets)
profile = ProfileReport(df, minimal=True)

# Custom title in the report
profile = ProfileReport(df, title="Titanic EDA Report", explorative=True)

# Treat a column as the target variable (highlighted in report)
profile = ProfileReport(df, explorative=True)

# Disable specific sections to speed up
profile = ProfileReport(df, correlations=None)

# For time-series data
profile = ProfileReport(df, tsmode=True, sortby="date_column")
```

---

### Line 3 — Saving the Report

```python
profile.to_file("output.html")
```

- Saves the entire interactive report as a **self-contained HTML file**.
- The file can be opened in any browser — no Python or internet connection needed.
- **Self-contained** means all CSS, JS, and chart data is embedded inside the HTML file itself.

#### Output methods available:

| Method | Output | Use case |
|--------|--------|---------|
| `profile.to_file("report.html")` | 📄 HTML file | Share with team, open in browser |
| `profile.to_notebook_iframe()` | 🖼️ Inline in Jupyter | View directly in notebook |
| `profile.to_json()` | `{}` JSON string | Programmatic access to all stats |
| `profile.to_widgets()` | 🔲 Jupyter widget | Interactive in notebook |

```python
# View inline inside the notebook (instead of saving to file):
profile.to_notebook_iframe()
```

### 📋 What the generated `output.html` report contains:

#### 1. Overview Tab
```
Dataset info:
  Number of variables    12
  Number of observations 891
  Missing cells          866 (8.1%)
  Duplicate rows         0
  Total size in memory   83.4 KiB
```

#### 2. Variables Tab (per column)
For each column you get:
- **Type detection**: Numeric / Categorical / Boolean / Date / Text / Unsupported
- **Missing values**: count + percentage
- **Unique values**: count + percentage
- **Distribution**: histogram or bar chart
- **Detailed stats**: mean, median, std, skewness, kurtosis (for numericals)
- **Most frequent values** (for categoricals)
- **Warnings**: flags like HIGH CARDINALITY, HIGH CORRELATION, MISSING, SKEWED, ZEROS

Example for `Age`:
```
Age  (Numeric)
  Distinct         88         Missing       177 (19.9%)
  Mean             29.7       Minimum       0.42
  Median           28.0       Maximum       80.0
  Std dev          14.5       Skewness      0.39
  ⚠ MISSING — 19.9% missing values
  ⚠ SKEWED — skewness = 0.39
```

#### 3. Interactions Tab
- Scatter plots for every pair of numerical columns.
- Helps spot linear/non-linear relationships between features.

#### 4. Correlations Tab
- **Pearson**: linear correlation between numerical columns.
- **Spearman**: rank-based correlation (handles non-linear relationships).
- **Cramér's V**: correlation between categorical columns.
- **Phik**: universal correlation measure (works across mixed types).
- Displayed as colour-coded heatmaps — instantly shows which features are related.

#### 5. Missing Values Tab
- **Bar chart**: how many values are missing in each column.
- **Matrix**: visual map of where NaNs are across all rows — helps spot patterns in missingness.
- **Heatmap**: correlation between missingness of different columns (are `Age` and `Cabin` missing together?).

#### 6. Sample Tab
- First 10 and last 10 rows of the raw data.

---

### 🔍 Data Analysis Context — Why this matters:

| Manual EDA | Automated with ProfileReport |
|------------|------------------------------|
| ~50 lines of code | 3 lines |
| Only covers what you remember to check | Comprehensive — checks everything |
| Static output | Interactive HTML — clickable, zoomable |
| Takes 30–60 minutes | Takes 30–60 seconds |
| Hard to share | Single `.html` file — open in any browser |

**The report directly drives your data preparation decisions:**

| Report finding | What you do next |
|---------------|-----------------|
| `Age` — 19.9% missing | Impute with median |
| `Cabin` — 77% missing | Drop the column |
| `Embarked` — 2 missing | Fill with mode (`'S'`) |
| `Fare` — skewness = 4.79 | Apply `np.log1p()` transform |
| `PassengerId` — HIGH CORRELATION with nothing | Drop — not a useful feature |
| `Pclass` ↔ `Fare` — high negative correlation | Don't use both; pick one or combine |

> 💡 **Best practice:** Run `ProfileReport` at the **very start** of every new dataset. The warnings and correlations it flags will guide your entire preprocessing and feature engineering pipeline.
