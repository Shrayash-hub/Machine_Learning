# Importing Libraries and Loading Data

**Explanation:** Pandas tabular data (CSV) ko load aur manipulate karne ke liye ek industry-standard library hai. `read_csv` function data ko Pandas DataFrame me convert karta hai.

**Context for ML:** Machine Learning models raw text ya CSV files ko direct nahi samajhte, unhe structured arrays aur matrices chahiye hote hain. Pandas data ko structure karta hai jisse humari poori ML pipeline (preprocessing, cleaning, scaling) implement karna easy ho jata hai.

```python
import pandas as pd
```

```python
df = pd.read_csv('train.csv')
```

# 1. How big is the data?

**Code used:** `df.shape`

**Explanation:** Yeh attribute humein ek tuple return karta hai `(number_of_rows, number_of_columns)` ke format me. Isse humein dataset ka volume pata chalta hai.

**Context for ML:** Dataset ka volume samajhna bohot zaroori hai model selection ke liye. Agar data bohot chota hai, toh complex Deep Learning models overfit kar jayenge, aur wahan traditional ML models (jaise Logistic Regression, SVM, Random Forest) better perform karenge. Is dataset me 891 rows hain, which is relatively very small.

```python
df.shape
```

# 2. How does the data look like?

**Code used:** `df.sample(5)`

**Explanation:** `sample(5)` randomly 5 rows utha kar dikhata hai. Log aksar `head()` use karte hain, par `sample()` better practice hai kyunki `head()` sirf shuruati rows dikhata hai jo kabhi-kabhi ek specific order me sorted ya biased ho sakti hain.

**Context for ML:** Ek quick glance me humein pata chal jata hai ki kis type ka data present hai—kya values categorical hain (jaise 'Sex' me male/female), kya string/text data hai (jaise 'Name', 'Ticket'), aur kya numerical hai. Ye observation aage chalkar Feature Engineering aur Encoding strategies decide karne me base set karti hai.

```python
df.sample(5)
```

# 3. What is the data type of cols?

**Code used:** `df.info()`

**Explanation:** Yeh function dataset ki memory footprint aur poori summary deta hai: total entries, har column ka naam, usme kitni non-null (not empty) values hain, aur specific column ka data type (`int`, `float`, ya `object`/string).

**Context for ML:** ML algorithms mathematically work karte hain, isliye strings (`object` type) direct feed nahi kiye ja sakte. `df.info()` dekh kar humein exactly pata chalta hai ki kin columns ko humein numerical format me encode karna padega (e.g., One-Hot Encoding for 'Embarked' ya Label Encoding for 'Sex').

```python
df.info()
```

# 4. Are there any missing values?

**Code used:** `df.isnull().sum()`

**Explanation:** `isnull()` har individual cell ko check karta hai ki wo khali (NaN) hai ya nahi (True/False return karta hai), aur `.sum()` un True values ko count kar leta hai column-wise.

**Context for ML:** Most ML libraries (jaise Scikit-Learn) missing values hone par training ke waqt error throw karti hain. Yeh step humein batata hai ki kahan imputation (filling missing values with mean/median/mode) ki zaroorat hai. Jaise yahan 'Age' me missing values hain jinhe hum average age se bhar sakte hain. 'Cabin' me bahut zyada (687) missing values hain, toh itna data fill karne se better hai hum us column ko drop hi kar dein.

```python
df.isnull().sum()
```

# 5. How does the data look mathematically?

**Code used:** `df.describe()`

**Explanation:** Yeh function sirf numerical columns ki deep statistical summary nikalta hai (count, mean, standard deviation, min, max, aur percentiles/quartiles - 25%, 50%, 75%).

**Context for ML:** Outlier detection ke liye yeh sabse initial aur best step hai! Jaise 'Fare' column me dhyaan se dekho: 75% logo ne sirf 31.00 tak pay kiya hai, but Max fare 512.32 hai. Ye ek bohot bada outlier hai jo model ki accuracy aur weights ko kharab kar sakta hai. Isse dekh kar hum decide karte hain ki data preprocessing step me Scaling (Standardization/Normalization) karna kitna zaroori hai.

```python
df.describe()
```

# 6. Are there duplicate values?

**Code used:** `df.duplicated().sum()`

**Explanation:** Yeh check karta hai ki kya dataset me koi aisi poori row hai jo exactly kisi aur row jesi (repeat) ho.

**Context for ML:** Duplicate rows data me bias create kar sakti hain, jisse hamara ML model un repeated patterns ko over-emphasize kar leta hai (memorize karne lagta hai). Data leakage bachane ke liye aur clean generalization ke liye duplicates ko hamesha remove karna chahiye.

```python
df.duplicated().sum()
```

# 7. How is the correlation between cols?

**Code used:** `df.corr(numeric_only=True)['Survived']`

**Fix Explanation:** Puraane Pandas versions me `df.corr()` automatically string/object columns ko ignore kar deta tha, but naye versions me ye strictly error deta hai (`could not convert string to float`) kyuki text ka math-based correlation nahi nikal sakta. Isliye `numeric_only=True` parameter add karna zaroori hai.

**Explanation:** Correlation (Pearson coefficient from -1 to 1) batata hai ki ek column ki value change hone par dusra linearly kaise behave karta hai.

**Context for ML:** Yeh **Feature Selection** ka sabse important tool hai. Yahan hum strictly dekh rahe hain ki 'Survived' (jo humara target/output variable hoga) ke sath baaki features ka kaisa relationship hai:
- 'Pclass' (-0.33) negatively correlated hai (Iska matlab, Pclass ki value jitni kam - yaani 1st Class - bachne ke chances utne zyada).
- 'Fare' (0.25) positively correlated hai (Yaani jisne ticket me zyada paise diye, unke bachne ka probability zyada tha).
Jo independent features target variable se lagbhag 0 correlated hote hain, hum unko model training se nikal (drop) sakte hain taaki model faster aur noise-free bane.

```python
df.corr(numeric_only=True)['Survived']
```
